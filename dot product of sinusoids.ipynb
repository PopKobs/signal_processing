Q2. Define a 2D array of dimensions 11x1000 and of type float.
Name this array S.  

Define a 2D array of dimensions 11x1000 and of type float.
Name this array C.  

For each ùëñth row in S, define a sine wave of frequency ùëì[ùëñ], magnitude 1, time
vector given by t.  

For each ùëñth row in C, define a cosine wave of frequency ùëì[ùëñ], magnitude 1,
time vector given by t.  

Plot all the sine waves defined in S in one figure.  

Plot all the cosine waves defined in C in another figure.

```python
num_rows = len(f)
num_cols = len(t)

S = np.zeros((num_rows, num_cols), dtype=float)
C = np.zeros((num_rows, num_cols), dtype=float)

for i in range(11):
    S[i, :] = np.sin(2 * np.pi * f[i] * t)

for i in range(11):
    C[i, :] = np.cos(2 * np.pi * f[i] * t)

plt.figure(figsize=(10, 6))
for i in range(11):
    plt.plot(t, S[i, :], label='sine wave of frequency {} Hz'.format(f[i]))
plt.title('Sine waves of various frequencies')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for i in range(11):
    plt.plot(t, C[i, :], label='cosine wave of frequency {} Hz'.format(f[i]))
plt.title('Cosine waves of various frequencies')
plt.legend()
plt.show()
```

Q3. For all the sine waves defined in S, dot product them with each other. Print your result by rounding off the values to 2 decimal places.

```python
dot_product_sin = np.dot(S, S.T)

print(np.round(dot_product_sin, 2))
```

Q4. For all the cosine waves defined in C, dot product them with each other.
Print your result by rounding off the values to 2 decimal places.

```python
dot_product_cos = np.dot(C, C.T)

print(np.round(dot_product_cos, 2))
```

Q5. Dot product all the sine waves in S with each of the cosine waves in C.
Print your result by rounding off the values to 2 decimal places.

```python
dot_product_C_S = np.dot(S, C.T)

print(np.round(dot_product_C_S, 2))
```

```python
sqrtSumSq = np.sqrt(dotSx**2 + dotCx**2)

plt.figure()
plt.stem(f, [1, 6, 3, 5, 7, 4, 2, 10, 8, 9, 0])
plt.title('Magnitude of each of the frequency components in x')
plt.show()
```

```python
plt.figure()
plt.stem(f, sqrtSumSq)
plt.title('sqrtSumSq')
plt.show()
```

Both plots are alike in the ratio of magnitude (y axis) of each bar relative to each other. Since the array is of length 1000, in the 2nd plot it makes sense that the y axis values are multiples of 1000. Both plots are hence identical.
