Q2. Read “tone2000Hz.wav”. Perform FFT on the audio signal. Plot the absolute value of the FFT results. Identify the index i.e. location of the two peaks in the FFT results from the figure plotted above. Compute the frequency bin size. Verify that the peaks happened at the correct frequencies in Hz as expected. 

```python
# samplerate, data = read("tone2000Hz.wav")
samplerate, data = read("2000.wav")
samples_for_six_seconds = 6 * samplerate

trimmed_data = data[:samples_for_six_seconds]

write("tone2000Hz.wav", samplerate, trimmed_data)
```

```python
duration = 6
fourier_freq = fftfreq(samplerate * duration, 1/samplerate)
fourier_freq
```

```python
y = fft(trimmed_data)

plt.figure(figsize=(10, 6))
plt.plot(np.abs(y))
plt.xlabel('Frequency (bins)')
plt.ylabel('Amplitude')
plt.title('Absolute Value of FFT Results')
plt.grid(True)
plt.show()
```

```python
len(y)
```

```python
peak_indices = np.where(np.abs(y) > 1e9)[0]  # Adjust threshold as needed
print("Peak Indices:", peak_indices)
```

```python
FreqBin = samplerate / len(y)
print("Frequency bin size:", FreqBin)

peak_frequencies = peak_indices * FreqBin
print("Peak Frequencies (Hz):", peak_frequencies)
```

The peak indices identified are [12000, 252600]. These correspond to the 12000th index from the left, & 12000th index from the right. After multiplying with the frequency bin size, we can obtain the correct peak frequencies [ 2000. 42100.]. A similar process is followed for 3000Hz, 5000Hz & 10000Hz.

Q2. Generate a chirp signal using the following formula: $y_{chirp} = sin(2\pi f t)$ where f = 500 + $\frac{10000 t}{t1 - t0}$  

Use a sampling frequency of 44.1kHz, duration 5 secs, start of chirp is at time t0=0. End of chirp is at time t1=5 seconds. Frequency at time t0 is 500Hz. Frequency at time t1 is 10000Hz.

```python
from scipy import signal
```

```python
fs = 44100  
t0 = 0  
t1 = 5  
f0 = 500  
f1 = 10500  
T = 5  

t = np.linspace(t0, t1, T*fs, False)

f = f0 + (f1-f0)*(t-t0)/(t1-t0)

y_chirp = np.sin(2*np.pi*f*t)

plt.figure()
plt.plot(t, y_chirp)
plt.title('Chirp Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()
```

Play the chirp signal using sd.play. Save the chirp signal into a file. Name the file “myChirp.wav”.

```python
sd.play(y_chirp, fs)

sd.wait()
y_chirp_normalized = np.int16(y_chirp/np.max(np.abs(y_chirp)) * 32767)
write("myChirp.wav", fs, y_chirp_normalized)
```

Read “myChirp.wav”. Use signal.spectrogram to create the spectrogram of the signal. Use plt.imshow to display the spectrogram. Save the figure into a file and name it Q3.

```python
fs, y_chirp = read("myChirp.wav")

frequencies, times, Sxx = signal.spectrogram(y_chirp, fs)

plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.colorbar(label='Intensity dB')
plt.title('Spectrogram of myChirp.wav')
plt.savefig('Q3.png')
plt.show()
```
